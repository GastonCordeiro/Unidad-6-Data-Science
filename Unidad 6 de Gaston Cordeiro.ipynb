4. Estime el puntaje z de las regiones
Para ello, modifique la función para generar dotplots de la semana pasada, donde deberá agregar
una opción al argumento statistic para calcular el puntaje z.

```python
def dotplot(df,plot_var,plot_by,global_stat=False,statistics="mean"):
    if global_stat:
        plt.axvline(np.mean(df)[plot_var])
        
    if statistics=="mean":
        grouped= df.groupby(plot_by)[plot_var].agg("mean")
    elif statistics=="median":   
        grouped= df.groupby(plot_by)[plot_var].agg("median")
    else:
        df["z"]= (df[plot_var]-df[plot_var].mean())/ df[plot_var].std()
        grouped= df.groupby(plot_by)["z"].agg("mean")
        plt.axvline(0)
        
    plt.plot(grouped.values, grouped.index, "o", color="grey")
    plt.title(plot_var)    
    
    plt.show()
```
